```python
# Abdullah Project-9


import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Replace this with the file upload or use your dataset.
data = {
    "A": [1, 2, 3, 4, 5],
    "B": [5, 4, 3, 2, 1],
    "C": [2, 3, 4, 5, 6],
    "D": [6, 5, 4, 3, 2]
}

df = pd.DataFrame(data)

# Function to display data and plots
def dashboard(selected_column, selected_value, x_column, y_column):
    # Filter the data
    filtered_df = df[df[selected_column] == selected_value]

    # Display the data preview
    print("Data Preview:")
    display(filtered_df.head())

    # Plot data
    plt.figure(figsize=(10, 6))
    plt.plot(filtered_df[x_column], filtered_df[y_column], marker='o')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title(f'{x_column} vs {y_column}')
    plt.grid(True)
    plt.show()

# Widgets for selection
column_selector = widgets.Dropdown(
    options=df.columns,
    value='A',  # default value
    description='Column:',
)

value_selector = widgets.Dropdown(
    options=df['A'].unique(),
    value=1,  # default value for filtering
    description='Value:',
)

x_selector = widgets.Dropdown(
    options=df.columns,
    value='A',  # default value for x-axis
    description='X-axis:',
)

y_selector = widgets.Dropdown(
    options=df.columns,
    value='B',  # default value for y-axis
    description='Y-axis:',
)

# Create an interactive dashboard
interact(dashboard,
         selected_column=column_selector,
         selected_value=value_selector,
         x_column=x_selector,
         y_column=y_selector);
```
